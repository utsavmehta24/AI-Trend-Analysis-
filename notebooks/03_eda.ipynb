```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../data/processed/ai_trends_cleaned.csv")
```

```python
df = df.sort_values(by="date")
df.head()
```

```python
plt.figure(figsize=(12, 5))
plt.plot(df['date'], df['interest'])
plt.title("Search Interest Trend for ChatGPT Over Time")
plt.xlabel("Year")
plt.ylabel("Search Interest")
plt.show()
```

```python
df.loc[df['interest'].idxmax()]
```

```python
df['year'] = pd.to_datetime(df['date']).dt.year
```

```python
yearly_avg = df.groupby('year')['interest'].mean()
yearly_avg
```

```python
plt.figure(figsize=(8, 5))
yearly_avg.plot(kind='bar')
plt.title("Average Yearly Search Interest")
plt.xlabel("Year")
plt.ylabel("Average Interest")
plt.show()
```
