#### Finding null values at exact row. 
1. df.isnull() return boolean value for each item if item is null then true otherwise false 
2. any(axis=1) tells that within the row is anything null ? if any coloum cotains null value then it assigns null as the whole result as a null value.

    col1 col2 col3 col4         result
eg. 1      as    n   as  ->     false
    2       a   dd   NaN ->     true

3. df finally takes out that row as seen in the result.

reset_index():

Imagine you deleted row #2. Your rows would be numbered 0, 1, 3, 4.

This command fixes that gap so they become 0, 1, 2, 3 again.

drop=True:

Without this: Pandas tries to be helpful and saves your old row numbers (0, 1, 3, 4) into a brand new column called "index".

With this: You are telling Pandas, "I don't need the old numbers. Just delete them."

```python
df = df.reset_index(drop=True)
```
